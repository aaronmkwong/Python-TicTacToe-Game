```python
# get libraries

import random 

# create functions used in main program

def display_board(board):
    '''Function to print board'''
    print ('-----------')
    print (' {} | {} | {}'.format(board[0],board[1],board[2]))
    print ('-----------')
    print (' {} | {} | {}'.format(board[3],board[4],board[5]))
    print ('-----------')
    print (' {} | {} | {}'.format(board[6],board[7],board[8]))
    print ('-----------')
    pass

def player_input():
    '''Function to take in player 1 and 2 marker (character) choices and return as a list;
    upper conversion before test as lower case will be converted for marker tuple;
    returns a list of player1 and player2 markers'''
    marker = ''
    # 
    while marker.upper() != 'X' and marker.upper() != 'O': 
        marker = input("Player 1 pick a marker 'X' or 'O': ")
    if marker.upper() == 'X': 
        return ['X','O']
    else:
        return ['O','X']
    pass

def place_marker(board, marker, position):
    '''Function to place marker on the board'''
    board[position-1] = marker
    return board
    pass

def win_check(board, mark):
    '''Function checks if a marker (character) has won;
    board numbers and positions are 
    horizontal (top, middle, bottom),
    vertical (left, middle, right),
    diagonal (left-right, right-left)'''
    return mark * 3 in [ board[0]+board[1]+board[2], board[3]+board[4]+board[5], board[6]+board[7]+board[8], \
                         board[0]+board[3]+board[6], board[1]+board[4]+board[7], board[2]+board[5]+board[8], \
                         board[0]+board[4]+board[8], board[2]+board[4]+board[6]    
                       ]
    pass

def choose_first():
    '''Function to randomly determine which player will go first'''
    return 'Player {} goes first'.format(random.choice([1,2]))
    pass

def space_check(board, position):
    '''function to determine whether a position is available'''
    return board[position] == ' '
    pass

def full_board_check(board):
    '''Function to determine whether the board is full'''
    return ' ' not in board
    pass

def player_choice(board):
    '''Function for user position input and if available return for later use'''
    choice = 0
    while choice not in range(1,len(board)+1) or not space_check(board, choice-1):
        try:
            choice = int(input('Choose an available position from 1 to {}: '.format(len(board))))
        except:
            continue
    return choice
    pass

def replay():
    '''Function to determine whether to play again'''
    play_again = ' '
    while  play_again.upper() not in  ['Y','N']:
        play_again = input('Would you like to play again? (Y/N): ')
    if play_again.upper() == 'Y':
        return True
    else:
        return False
    pass

# main program

# greeting
print('\033[1m'+'Welcome to Tic Tac Toe!'+'\033[0m'+'\n')
board = [' ']*9
display_board(board)

# notify users of board positions
print('\n'+'Before starting the game note the board positions...'+'\n')
positions = [1,2,3,4,5,6,7,8,9]
display_board(positions)
print('\033[1m'+'\n'+'Get ready to play!...'+'\033[0m'+'\n')
    
# obtain markers for player 1 and 2
mkrs = player_input()
print('\n'+'Thanks!...'+'\n')

# let users know who goes first
strt_plr_message = choose_first()
strt_plr_no = int(strt_plr_message[7])
strt_plr_mkr = mkrs[strt_plr_no-1]
print(strt_plr_message + ' and plays ' + "'" + strt_plr_mkr + "'."+'\n')

# initialize loop for current player
no_plrs = [1,2]  
curr_plr_no = strt_plr_no
curr_plr_mkr = strt_plr_mkr

while full_board_check(board) == False or win_check(board, curr_plr_mkr) == False:
    
    print( 'Player '+ str(curr_plr_no) )
    mkr_pos = player_choice(board)
    place_marker(board, curr_plr_mkr, mkr_pos)
    display_board(board)
    print('\n')
    if win_check(board, curr_plr_mkr) == True:
        # declare a winner
        print( '\033[1m'+'Player '+str(curr_plr_no)+' wins!'+'\033[0m'+'\n')
        # ask to play again
        if replay() == True:
            # reset loop variables, prompt for markers, randomly determine who goes first
            board = [' ']*9
            print('\033[1m'+'\n'+'Get ready to play!...'+'\033[0m'+'\n')
            mkrs = player_input()
            print('\n'+'Thanks!...'+'\n')
            strt_plr_message = choose_first()
            strt_plr_no = int(strt_plr_message[7])
            strt_plr_mkr = mkrs[strt_plr_no-1]
            print(strt_plr_message + ' and plays ' + "'" + strt_plr_mkr + "'."+'\n')
            curr_plr_no = strt_plr_no
            curr_plr_mkr = strt_plr_mkr
            continue
        else:
            # say goodbye
            print('\n'+'\033[1m'+'Thanks for playing. Goodbye!')
            break
    elif full_board_check(board) == True:
        # declare a draw
        print( '\033[1m'+'The game is a draw!'+'\033[0m'+'\n')
        # ask to play again
        if replay() == True:
            # reset loop variables, prompt for markers, randomly determine who goes first 
            board = [' ']*9
            print('\033[1m'+'\n'+'Get ready to play!...'+'\033[0m'+'\n')
            mkrs = player_input()
            print('\n'+'Thanks!...'+'\n')
            strt_plr_message = choose_first()
            strt_plr_no = int(strt_plr_message[7])
            strt_plr_mkr = mkrs[strt_plr_no-1]
            print(strt_plr_message + ' and plays ' + "'" + strt_plr_mkr + "'."+'\n')
            curr_plr_no = strt_plr_no
            curr_plr_mkr = strt_plr_mkr
            continue
        else:
            # say goodbye
            print('\n'+'\033[1m'+'Thanks for playing. Goodbye!')
            break
    else: 
        # set current player to the other number and marker 
        curr_plr_no = [x for x in no_plrs if no_plrs.index(x) != no_plrs.index(curr_plr_no)].pop()
        curr_plr_mkr = mkrs[curr_plr_no-1]
              
```
